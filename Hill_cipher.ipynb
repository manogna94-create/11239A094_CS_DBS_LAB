```python
import numpy as np

def main():
    a = np.array([[6, 24, 1], [13, 16, 10], [20, 17, 15]], dtype=np.uint32)
    b = np.array([[8, 5, 10], [21, 8, 21], [21, 12, 8]], dtype=np.uint32)
    c = [0] * 20
    d = [0] * 20

    msg = input("Enter plain text: ")

    # Convert plain text characters to numerical values (0-25)
    for i in range(len(msg)):
        c[i] = ord(msg[i]) - 65
        print(c[i], end=" ")

    # Encryption process (Only uses the first 3 characters of 'c' due to range(3))
    for i in range(3):
        t = 0
        for j in range(3):
            t += a[i][j] * c[j]
        d[i] = t % 26

    print("\nEncrypted Cipher Text :", end="")
    for i in range(3):
        print(" ", chr(d[i] + 65), end="")

    # Decryption process (Only uses the first 3 characters of 'd' due to range(3))
    for i in range(3):
        t = 0
        for j in range(3):
            t += b[i][j] * d[j]
        c[i] = t % 26

    print("\nDecrypted Cipher Text :", end="")
    for i in range(3):
        print(" ", chr(c[i] + 65), end="")

    print()

if __name__ == "__main__":
    main()
```
