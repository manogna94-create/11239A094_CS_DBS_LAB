```python
import math
p, q, n, t = 0, 0, 0, 0
flag = 0
e, d, temp, m, en = [], [], [], [], []
msg = ""

def prime(pr):
    j = int(math.sqrt(pr))
    for i in range(2, j + 1):
        if pr % i == 0:
            return 0
    return 1

def ce():
    global e, d, t, p, q
    k = 0
    for i in range(2, t):
        if t % i == 0:
            continue

        flag = prime(i)
        if flag == 1 and i != p and i != q:
            e.append(i)
            flag = cd(e[k])

            if flag > 0:
                d.append(flag)
            k += 1
            if k == 99:
                break

def cd(x):
    k = 1
    while True:
        k = k + t
        if k % x == 0:
            return k // x

def encrypt():
    global msg, m, n, e, en, temp
    key = e[0]

    for i, char in enumerate(msg):
        pt = ord(char) - 96

        k = pow(pt, key, n)

        temp.append(k)

        ct = k + 96

        en.append(ct)

    en.append(-1)

    print("\nTHE ENCRYPTED MESSAGE IS")
    print(''.join(chr(c) for c in en[:-1]))

def decrypt():
    global temp, n, d, m
    key = d[0]
    i = 0

    while en[i] != -1:
        ct = temp[i]

        k = pow(ct, key, n)
        pt = k + 96

        m.append(pt)
        i += 1
    m.append(-1)

    print("\nTHE DECRYPTED MESSAGE IS")
    print(''.join(chr(c) for c in m[:-1]))

def main():
    global p, q, n, t, msg, m

    print("\nENTER FIRST PRIME NUMBER")
    p = int(input())

    flag = prime(p)
    if flag == 0:
        print("\nWRONG INPUT")
        return

    print("\nENTER ANOTHER PRIME NUMBER")
    q = int(input())

    flag = prime(q)
    if flag == 0 or p == q:
        print("\nWRONG INPUT")
        return

    print("\nENTER MESSAGE")
    msg = input()
    m = [ord(c) for c in msg]

    n = p * q
    t = (p - 1) * (q - 1)

    ce()

    print("\nPOSSIBLE VALUES OF e AND d ARE")
    for i in range(len(e)):
        print(f"\n{e[i]}\t{d[i]}")

    encrypt()
    decrypt()

if __name__ == "__main__":
    main()
```
